```python
import pymysql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
plt.rcParams['font.family']='Malgun Gothic'
%matplotlib inline
```

```python
data_popul=pd.read_excel('data/2019popul.xlsx')
data_popul=data_popul.iloc[1:,1:]
data_popul.columns=data_popul.iloc[0,:].tolist()
data_popul=data_popul.iloc[1:,:]
data_popul.to_pickle('data/2019popul.pkl')
data_popul
```

```python
data_popul=data_popul[data_popul['행정기관'].apply(lambda x: re.match('\w+ \w+',x) and True or False)]
data_popul=data_popul[data_popul['행정기관'].apply(lambda x: False if re.match('\w+ \w+ \w+',x) else True)]
data_popul=data_popul.sort_values('행정기관')
data_popul=data_popul.loc[:,['행정기관','총인구수']]
```

```python
data_popul.to_pickle('data/시군구_인구수.pkl')
```
