```python
def hist_data(code):
    try:
        conn_=pymysql.connect(
        host='khk.cepsu2i8bkn5.ap-northeast-2.rds.amazonaws.com',
        user='khk',
        password='k2hyokwang2!',
        charset='utf8mb4',
        cursorclass=pymysql.cursors.DictCursor,
        port=3306,
        db='hist_data'
        )

        with conn_.cursor() as cursor:
            sql=f'SELECT * FROM `{code}`;'
            cursor.execute(sql)
            db_table=cursor.fetchall()
            data=pd.DataFrame(db_table)

    except  Exception as e:
        print(e)
    finally:
        conn_.close()
        data.index=data['날짜']
        del data['날짜']
        data=data[['종가']]
        data['종가']=data['종가'].apply(lambda x: re.sub(',','',x))
        data=data.astype(np.float64)
        data['수익률']=round(data['종가']/data['종가'].shift(1)-1,4)
        data=data.iloc[1:,:]['수익률']
    return data
```

```python
def get_corr(hist1,hist2):
    data1=hist_data(hist1)
    data2=hist_data(hist2)
    data=pd.merge(data1,data2,how='inner',left_index=True,right_index=True)
    data.columns=[hist1,hist2]
    data=data.fillna(0)
    r=data.corr(method='pearson')
    r=r.iloc[0,1]
    r=round(r,4)
    return r
```

```python
corp_code=[x.get('종목코드') for x in rows_]

matx_frame=np.zeros(len(corp_code)**2).reshape(len(corp_code),len(corp_code))
matx=pd.DataFrame(matx_frame,index=corp_code,columns=corp_code)
# matx=matx.iloc[:10,:10]
matx=matx.astype(np.float)
matx
```

```python
for n,i in enumerate(matx.index):
    for j in tqdm(matx.columns[n:]):
        matx.loc[i,j]=get_corr(i,j)
for n,i in tqdm(enumerate(matx.columns)):
    for j in tqdm(matx.index[n:]):
        matx.loc[j,i]=matx.loc[i,j]
```
